```python
colors=['blue','#Ffcc00','red',"green"]
```

```python
sizes = [10,10,10,10]
internal=0.9
external=0.05
ps=[[internal,external, external, external],
    [external,internal, external, external],
    [external,external, internal, external],
    [external,external, external, internal]
]
g = nx.stochastic_block_model(sizes,ps)
groups=[0]*10+[1]*10+[2]*10+[3]*10
nx.set_node_attributes(g,{i:groups[i] for i in range(40)},"group")
```

```python
pp = plot_interactive(g,graph_size=400,communities="group",node_size=10,labels=False)
pp.show("pouet.html")
display(HTML("pouet.html"))

df = nx.to_pandas_adjacency(g)
sns.set(rc={'figure.figsize':(6,6)})
ax = sns.heatmap(ps, annot=True,cmap="YlGnBu",center=0.5,vmin=0, vmax=1,cbar=False,annot_kws={"fontsize":20}
)
for color,tick in zip(colors,ax.yaxis.get_major_ticks()):
    tick.label1.set_color(color) #set the color property
for color,tick in zip(colors,ax.xaxis.get_major_ticks()):
    tick.label1.set_color(color) #set the color property
```

```python
import numpy as np
import pandas as pd
```

```python
g=nx.watts_strogatz_graph(20,4,0)
plt.figure(figsize=(8,8))

pos = nx.circular_layout(g)
#pos={x:v*0.6 for x,v in pos.items()}

ax = plt.gca()
plt.margins(0.05)
# Draw nodes and edges
#nx.draw_networkx_edges(
#    g, pos,
#    connectionstyle="arc3,rad=0.1")#pp.show("test.html")
#display(HTML("test.html"))
for edge in g.edges():
    source, target = edge
    rad = 0.20
    #c = edge in elarge
    arrowprops=dict(arrowstyle="-", 
                    color='grey',
                    connectionstyle=f"arc3,rad={rad}",
                    linestyle= '-',
                    alpha=0.3,
                    linewidth=2)
    ax.annotate("",
                xy=pos[source],
                xytext=pos[target],
                arrowprops=arrowprops
               )
nx.draw_networkx_nodes(g, pos, node_size=30, node_color='black')
```

```python
g=nx.watts_strogatz_graph(20,4,0.3)

plt.figure(figsize=(8,8))

pos = nx.circular_layout(g)
#pos={x:v*0.6 for x,v in pos.items()}

ax = plt.gca()
plt.margins(0.05)

for edge in g.edges():
    source, target = edge
    rad = 0.3
    c= abs(source-target) <=2 or abs(source-target)>=len(g.nodes)-2
    arrowprops=dict(arrowstyle="-", 
                    color='grey' if c else 'blue',
                    connectionstyle=f"arc3,rad={rad}",
                    linestyle= '-',
                    alpha=0.3,
                    linewidth=2)
    ax.annotate("",
                xy=pos[source],
                xytext=pos[target],
                arrowprops=arrowprops
               )
nx.draw_networkx_nodes(g, pos, node_size=30, node_color='black')
```

```python
records = []
for p in np.logspace(-3, 0, num=20):
  for r in range(100):
    print(p)
    g=nx.watts_strogatz_graph(200,6,p)
    records.append((p,nx.transitivity(g),nx.average_shortest_path_length(g)))
```

```python
df =pd.DataFrame.from_records(records,columns=["p","C","d"])
```

```python
df["C"]=df["C"]/records[0][1]
df["d"]=df["d"]/records[0][2]
```

```python
df2 = df.groupby("p").mean()
```
